```python
import sys
import os
import datetime

# Get absolute path to the 'src' directory (one level up from 'notebooks')
sys.path.append(os.path.abspath(os.path.join('..', 'src')))

from gcal_ai_agent.tools import get_calendar_service
```

```python
max_results: int = 10
```

```python
service = get_calendar_service()
```

```python
# Get current time in ISO format (required by the API)
# The modern, timezone-aware way (Python 3.11+)
now = datetime.datetime.now(datetime.UTC).isoformat().replace('+00:00', 'Z')
now
```

```python
events_result = service.events().list(
    calendarId='primary', 
    timeMin=now,
    maxResults=max_results, 
    singleEvents=True,
    orderBy='startTime'
).execute()
```

```python
events = events_result.get('items', [])
events
```

```python
if not events:
    print("No upcoming events found.")
```

```python
# Format the events into a readable string for the LLM
output = []
for event in events:
    start = event['start'].get('dateTime', event['start'].get('date'))
    summary = event.get('summary', 'No Title')
    output.append(f"{start} - {summary}")
```
