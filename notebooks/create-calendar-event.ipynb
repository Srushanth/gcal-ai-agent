```python
import sys
import os
import datetime

# Get absolute path to the 'src' directory (one level up from 'notebooks')
sys.path.append(os.path.abspath(os.path.join('..', 'src')))

from gcal_ai_agent.tools import get_calendar_service
```

```python
datetime.datetime.now(datetime.UTC).isoformat().replace('+00:00', 'Z')
```

```python
summary: str = "Test summary"
start_time: str = None
end_time: str = None
```

```python
# 1. Get the current local time with timezone info
now = datetime.datetime.now().astimezone()
# 2. Add time using timedelta
start_time = now + datetime.timedelta(hours=1)  # Starts 1 hour from now
end_time = start_time + datetime.timedelta(hours=1)  # Ends 1 hour after start

start_time = start_time.isoformat()
end_time = end_time.isoformat()

(start_time, end_time)
```

```python
service = get_calendar_service()
```

```python
event_payload = {
    'summary': summary,
    'start': {
        'dateTime': start_time,
        'timeZone': 'UTC', # Adjust to your preferred default timezone
    },
    'end': {
        'dateTime': end_time,
        'timeZone': 'UTC',
    }
}
event_payload
```

```python
event = service.events().insert(
    calendarId='primary', 
    body=event_payload
).execute()
```
